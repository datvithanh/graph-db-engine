```python
di = {
    'elder': ('Old', 'OldStatus', 'Người già'),
    'teen': ('Teen', 'TeenStatus', 'Thanh thiếu niên'),
    'mature': ('Mature', 'MatureStatus', 'Người trưởng thành'),
    'child': ('Child', 'ChildStatus', 'Trẻ em'),
    'pregnant': ('Pregnant', 'PregnantStatus', 'Phụ nữ mang thai')
}

df = df.set_index('name_enfood')
```

```python
foods = df.index
```

```python
foods = list(foods)
foods
```

```python
for k,v in di.items():
    # comment the following line to import
    break
    consumer_group = ConsumerGroup(name_en = k, name_vi = v[-1]).save()
    for food in foods:
        if not (pd.isna(df[v[1]][food]) or pd.isna(df[v[0]][food])):
            print(df[v[1]][food].lower(), df[v[0]][food])
            try:
                food_node = Food.nodes.get(name_en=food)
            except:
                food_node = Food(name_en = 'apple', name_vi = 'Táo').save()
            food_node.consumer_groups.connect(consumer_group, {'description': df[v[0]][food], 'status': df[v[1]][food]})
```
